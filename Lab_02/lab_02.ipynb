## Copying file form Original data to Expriment_small_data / train

```python
import shutil
import os

# Path for from where to copy files
source_folder = "../Lab_02/Original_data/train/train"

# Path to where files going to be copy
target_folder = "../Lab_02/expriment_small_data/train"

# How many file is going to be copy for each category
files_per_category = 800

total_count = 0


# Loop through the files in the source folder and MOVE the required number of files for each category
for filename in os.listdir(source_folder):
    # Check if the filename indicates a "cat" or "dog" image
    if filename.startswith("cat") or filename.startswith("dog"):
        source_path = os.path.join(source_folder, filename)
        target_path = os.path.join(target_folder, filename)
        shutil.move(source_path, target_path)
        total_count += 1

    # If the required number of total files has been MOVED, exit the loop
    if total_count >= 2 * files_per_category:
        break

print("Files copied succesfully")
```

## c) Resize so that all images are the same size and save them in numpy arrays with the following structure:

```python
import os
import numpy as np
from PIL import Image

def resize_image(im, new_width, new_height):
    resized_image = im.resize((new_width, new_height))
    return np.array(resized_image)

original_images_path = "../Lab_02/expriment_tiny_data/train"
resized_images_path = "../Lab_02/expriment_tiny_data/resized_numpy"

# Create a new folder to save resized images if it doesn't exist
if not os.path.exists(resized_images_path):
    os.makedirs(resized_images_path)

resized_images = []

for file in os.listdir(original_images_path):
    if file.endswith(".jpg"):
        file_path = os.path.join(original_images_path, file)
        im = Image.open(file_path)
        resized_image = resize_image(im, 100, 100)  # Resize images to 100x100 pixels
        resized_images.append(resized_image)

# Stack resized images along the first axis to create the sample dimension
resized_images_array = np.stack(resized_images)

# Save numpy array to file
np.save("resized_images.npy", resized_images_array)

print("Resized images saved as numpy array!")
print(f"Shape of resized images array:, {resized_images_array.shape}")
```

```python
from pathlib import Path
import numpy as np
import cv2
```

```python
def load_images_from_folder(folder, target_size):
    folder_path = Path(folder)
    if not folder_path.exists():
        print(f"The folder {folder} does not exist.")
        return np.array([]), np.array([])

    images = []
    labels = []
    
    for class_label, class_folder in enumerate(['cat', 'dog']):
        for image_path in folder_path.glob(f"{class_folder}.*.jpg"):
            print(f"Loading image: {image_path}")
            image = cv2.imread(str(image_path))
            if image is None:
                print(f"Failed to load image: {image_path}")
                continue
            image_resized = cv2.resize(image, target_size)
            images.append(image_resized)
            labels.append(class_label)

    return np.array(images), np.array(labels)

target_size = (250, 250)  

# Load small dataset
X_small_train, y_small_train = load_images_from_folder("../Lab_02/expriment_small_data/train", target_size)
X_small_val, y_small_val = load_images_from_folder("../Lab_02/expriment_small_data/val", target_size)
X_small_test, y_small_test = load_images_from_folder("../Lab_02/expriment_small_data/test", target_size)

print(f'Number of small train dataset images: {len(X_small_train)}')
print(f'Number of small val dataset images: {len(X_small_val)}')
print(f'Number of small test dataset images: {len(X_small_test)}')
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
```

# Convolutional Neural Network (CNN)

```python
def create_model(input_shape):
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        layers.MaxPooling2D((2, 2)),
        
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        
        layers.Conv2D(256, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        
        layers.Flatten(),
        layers.Dense(512, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(1, activation='sigmoid')
    ])
    
    return model
```

# Model Training 

```python
# Load the data
X_train, y_train = X_small_train, y_small_train
X_val, y_val = X_small_val, y_small_val
X_test, y_test = X_small_test, y_small_test

# Normalize pixel values to be between 0 and 1
X_train, X_val, X_test = X_train / 255.0, X_val / 255.0, X_test / 255.0

# Reshape data to fit the model
input_shape = X_train.shape[1:]

# Create the model
model = create_model(input_shape)

# Compile the model
model.compile(optimizer='adam',
            loss='binary_crossentropy',
            metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_val, y_val))
```

# Evaluation of Model 

```python
# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```
