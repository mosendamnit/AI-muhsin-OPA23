# b) Load 10 images randomly, plot them and extract their labels and print in the title.

```python
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Path to the folder containing your images
folder_path = "../Lab_02/train/train"

# List all files in the folder
image_files = [file for file in os.listdir(folder_path) if file.endswith('.jpg') or file.endswith('.png')]

# Pick 10 random image files
random_image_files = random.sample(image_files, 10)

# Plot the 10 random images
fig, axes = plt.subplots(2, 5, figsize=(15, 6))

for i, ax in enumerate(axes.flat):
    # Load random image
    random_image_file = random_image_files[i]
    random_image_path = os.path.join(folder_path, random_image_file)
    random_image = mpimg.imread(random_image_path)
    
    # Extract label (assuming label is the first part of the filename before underscore)
    label = random_image_file.split('_')[0]
    
    # Plot the image
    ax.imshow(random_image)
    ax.axis('off')
    ax.set_title(label)  # Set title as the label

plt.show()
```

# Create the folder structure below using Python:

## Copying file form Original data to Expriment_small_data / train

```python
import shutil
import os

# Path for from where to copy files
source_folder = "../Lab_02/Original_data/train/train"

# Path to where files going to be copy
target_folder = "../Lab_02/expriment_small_data/train"

# How many file is going to be copy for each category
files_per_category = 800

total_count = 0


# Loop through the files in the source folder and copy the required number of files for each category
for filename in os.listdir(source_folder):
    # Check if the filename indicates a "cat" or "dog" image
    if filename.startswith("cat") or filename.startswith("dog"):
        source_path = os.path.join(source_folder, filename)
        target_path = os.path.join(target_folder, filename)
        shutil.copy2(source_path, target_path)
        total_count += 1

    # If the required number of total files has been copied, exit the loop
    if total_count >= 2 * files_per_category:
        break

print("Files copied succesfully")
```
