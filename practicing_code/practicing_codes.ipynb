```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Original feature (single feature)
X = np.array([[2], [3], [5]])  # Example data: house sizes
y = np.array([6, 9, 15])  # Example target: house prices

# Create polynomial features (up to degree 3)
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)

# Linear regression on original feature
model_linear = LinearRegression()
model_linear.fit(X, y)
y_linear_pred = model_linear.predict(X)

# Linear regression on polynomial feature (x^2)
model_poly = LinearRegression()
model_poly.fit(X_poly[:, 1:2], y)  # Use only x^2 feature
y_poly_pred = model_poly.predict(X_poly[:, 1:2])

# Plotting
plt.figure(figsize=(12, 5))

# Scatter plot of original data
plt.subplot(1, 2, 1)
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_linear_pred, color='red', label='Linear Fit')
plt.xlabel('House Size')
plt.ylabel('House Price')
plt.title('Linear Regression (Single Feature)')
plt.legend()

# Scatter plot of polynomial feature (x^2)
plt.subplot(1, 2, 2)
plt.scatter(X_poly[:, 1], y, color='blue', label='Data')
plt.plot(X_poly[:, 1], y_poly_pred, color='green', label='Linear Fit (x^2)')
plt.xlabel('House Size (Squared)')
plt.ylabel('House Price')
plt.title('Linear Regression (Polynomial Feature: x^2)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Sample data
age = np.array([20, 30, 40, 50, 60, 70, 80])
income = np.array([20, 40, 60, 80, 100, 120, 200])

# Min-Max Scaling
age_scaled = (age - age.min()) / (age.max() - age.min())
income_scaled = (income - income.min()) / (income.max() - income.min())

# Standardization
age_mean = age.mean()
age_std = age.std()
income_mean = income.mean()
income_std = income.std()

age_standardized = (age - age_mean) / age_std
income_standardized = (income - income_mean) / income_std

print("Age (original):", age)
print("Income (original):", income)
print("Age (scaled):", age_scaled)
print("Income (scaled):", income_scaled)
print("Age (standardized):", age_standardized)
print("Income (standardized):", income_standardized)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of car samples
samples = 10

# Generate ages of cars (in years) between 0 and 10
age = np.random.uniform(0, 10, (samples, 1))

# Define a function to calculate car prices based on age
def calculate_price(age):
    return age**2 + 0.5*age + 3  # Price = age^2 + 0.5*age + 3

# Calculate prices for each age with some random noise
price = calculate_price(age) + np.random.randn(samples, 1)

# Generate ages for plotting
x = np.linspace(0, 10, samples)

# Plot the car ages and prices
plt.plot(age, price, '.', label='Car Data')
plt.plot(x, calculate_price(x), label='Original Price Function')
plt.xlabel('Age (Years)')
plt.ylabel('Price ($)')
plt.legend()
plt.title('Car Price Prediction')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate synthetic data with a nonlinear relationship
X = np.linspace(-3, 3, 50)
y = X**2 + np.random.normal(0, 1, 50)  # Quadratic relationship with some noise

# Reshape X to a column vector
X = X.reshape(-1, 1)

# Fit a linear regression model
model = LinearRegression()
model.fit(X, y)

# Plot the data points
plt.scatter(X, y, color='blue', label='Data')

# Plot the predicted values from the linear regression model
plt.plot(X, model.predict(X), color='red', label='Linear Regression')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Underfitting Example')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

np.random.seed(0)
x = np.linspace(-3, 3, 100)
y = x**2 + np.random.normal(0, 1, 100)

x = x.reshape(-1, 1)

x_train, x_test = x[:70], x[70:]
y_train, y_test = y[:70], y[70:]

poly = PolynomialFeatures(degree=30)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)

model = LinearRegression()
model.fit(x_train_poly, y_train)

y_train_pred = model.predict(x_train_poly)
y_test_pred = model.predict(x_test_poly)

plt.scatter(x, y, color='blue')  # Fix color to match scatter points with the line
plt.plot(x, model.predict(poly.transform(x)), color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression (Degree=30)')
plt.show()
```
